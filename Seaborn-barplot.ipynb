```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.barplot(x='day', y='tip', data=tips)
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

```python
sns.barplot(x='day', y='tip', data=tips, hue='sex', palette='Spectral')
```

```python
sns.barplot(x='total_bill', y='day', data=tips, palette='spring')
```

```python
from numpy import median
sns.barplot(x='day', y='total_bill', data=tips, estimator=median)
```

```python
sns.barplot(x='smoker', y='tip', data=tips)
```

```python
sns.barplot(x='smoker', y='tip', data=tips, ci=99)
```

```python
sns.barplot(x='day', y='total_bill', data=tips, palette='spring', capsize=0.3)
```

```python
sns.barplot(x='day', y='total_bill', data=tips, palette='spring', saturation=0.3)
```
